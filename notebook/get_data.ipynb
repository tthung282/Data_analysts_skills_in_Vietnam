```python
# import base64
import pandas as pd
from serpapi import GoogleSearch
import datetime
import streamlit as st
```

```python
search_term = "data analyst"
search_location = ['Vietnam','Ho Chi Minh City, Ho Chi Minh City, Vietnam','Hanoi,Hanoi,Vietnam','Da Nang, Da Nang, Vietnam']

for i in range (4):
    for num in range(45):

        start = num *10
        params = {
            "api_key": "f124ea12d9bde1b9a25bb15d0619afa425fab97a70758057b10713a1a527cbf3", #Fill in with your API key from SerpApi
            "engine": "google_jobs",
            "google_domain": "google.com.vn",
            "q": search_term,
            "hl": "vi",
            "gl": "vn",
            "location": search_location[i],
            "chips": "date_posted:today",
            "start": start,
        }

        search = GoogleSearch(params)
        results = search.get_dict()

        # check if the last search page (i.e., no results)
        try:
            if results['error'] == "Google hasn't returned any results for this query.":
                break
        except KeyError:
            print(f"Getting SerpAPI data for page: {start}")
        else:
            continue

        # create dataframe of 10 pulled results
        jobs = results['jobs_results']
        jobs = pd.DataFrame(jobs)
        jobs = pd.concat([pd.DataFrame(jobs), 
                        pd.json_normalize(jobs['detected_extensions'])], 
                        axis=1).drop('detected_extensions', 1)
        jobs['date_time'] = datetime.datetime.now()

        # concat dataframe
        if start == 0:
                jobs_all1 = jobs
        else:
                jobs_all1 = pd.concat([jobs_all1, jobs])

        jobs_all1['search_term'] = search_term
        jobs_all1['search_location'] = search_location[i]

    if i == 0:
            jobs_all = jobs_all1
    else:
            jobs_all = pd.concat([jobs_all, jobs_all1])


jobs_old = pd.read_csv('C:/Users/admin/Python_analysis/Streamlit/Data_analysts_skills_in_Vietnam/data6.csv')

jobs_all = pd.concat([jobs_all,jobs_old])

jobs_all.to_csv('C:/Users/admin/Python_analysis/Streamlit/Data_analysts_skills_in_Vietnam/data6.csv')
pd.DataFrame(jobs_all)
```
